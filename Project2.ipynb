```python
# Now again importing train_test_split for the decision tree and random forest since..the standard scaler is not for these both algorithms
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80)
```

```python
x_train
```

```python
x_test
```

```python
# Decision tree
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dct_model=DecisionTreeClassifier()
```

```python
dct_model.fit(x_train,y_train)
```

```python
ans4=dct_model.predict(x_test)
ans4
```

```python
a4=dct_model.score(x_test,y_test)
print("The accuracy of the decision tree classifier is",a4)
```

```python
#Random forest algorithm
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc_model=RandomForestClassifier(n_estimators=7)
```

```python
rfc_model.fit(x_train,y_train)
```

```python
ans5=rfc_model.predict(x_test)
ans5
```

```python
a5=rfc_model.score(x_test,y_test)
print("The accuracy of the Random forest classifier is",a5)
```

```python
# Now comparing the algorithms->
```

```python
print("The accuracy of the support vector machine is",a1)
print("The accuracy of the Logistic Regression is",a2)
print("The accuracy of the knn is",a3)
print("The accuracy of the decision tree classifier is",a4)
print("The accuracy of the Random forest classifier is",a5)
```

```python
# So among the 5 algorithms, the highest accuracy is gained from randomforest
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(ans5,y_test)
cm
```

```python
from sklearn.metrics import classification_report
```

```python
print("Classification Report:")
print(classification_report(ans5, y_test))
```

```python
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Pred 0', 'Pred 1'], 
            yticklabels=['Actual 0', 'Actual 1'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

```python
from sklearn.metrics import roc_curve, auc

# Get the predicted probabilities (not just the final prediction)
y_prob = rfc_model.predict_proba(x_test)[:, 1]  # Probabilities for class 1

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate (Recall)')
plt.legend(loc='lower right')
plt.show()
```

```python
#The Random Forest model performs well with an accuracy of 72% and an AUC score of 0.77. While accuracy is a useful metric, the ROC AUC provides a more reliable evaluation of the model, especially in cases of class imbalance.

```
